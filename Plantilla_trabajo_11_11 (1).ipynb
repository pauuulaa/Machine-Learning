```python
y = np.array([10,12,13,15,13,25,35,34,40,41])
```

```python
plt.scatter(x, y)
#plt.plot(x, larecta, color = 'red')
```

```python
m = 0
c = 0
```

```python
y_hat = m*x + c
```

```python
plt.scatter(x, y)
plt.plot(x, y_hat, color = 'red')
```

```python
1/len(x) * sum((y - y_hat)**2)  
```

```python
# derivada parcial cobre m y derivada parcial sobre c
dm = 2/len(x) * sum((y - y_hat)*(-x))
dc = (2/len(x)) * sum((y - y_hat)*(-1))
```

```python
m = m - dm*0.0001
c = c - dc*0.0001
```

```python
m,c
```

```python
for i in range(1000):
    y_hat = m*x + c
    dm = 2/len(x) * sum((y - y_hat)*(-x))
    dc = (2/len(x)) * sum((y - y_hat)*(-1))
    m = m - dm*0.0001
    c = c - dc*0.0001
```

```python
m,c
```

```python
plt.scatter(x, y)
plt.plot(x, m*x + c, color = 'red')
```
